```
import pandas as pd

df = pd.read_csv('/content/bank.csv',sep=";")
```

```
print(df)
target=df['y']
print(target)
Y = set()
for val in target:
    Y.add(val)
Y = list(Y)

print(df)
```

```
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
#from sklearn.cross_validation import train_test_split

X = df.values.tolist()
Y=target
x_train = []
y_train = []
x_test = []
y_test = []

x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.8,shuffle=True)
```

```
print(x_train)
print(y_train)
print(x_test)
print(y_test)
```

```
import numpy as np
x_train = np.array(x_train)
y_train = np.array(y_train)
x_test = np.array(x_test)
y_test = np.array(y_test)
```

```
y_train = y_train.reshape(3616,1)
y_test = y_test.reshape(905,1)
```

```
from sklearn.preprocessing import LabelEncoder
Le = LabelEncoder()
for i in range(0, 3616):
  x_train[i]= Le.fit_transform(x_train[i])
for i in range(0, 905):
  x_test[i] = Le.fit_transform(x_test[i])
```

```
for i in range(0, 3616):
  y_train[i]= Le.transform(y_train[i])
for i in range(0, 905):
  y_test[i] = Le.transform(y_test[i])
```

```
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
ohe.fit(x_train)
x_train = ohe.transform(x_train)
x_test = ohe.transform(x_test)
```

```
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test)
```

```
y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```
from sklearn import svm
SVM = svm.SVC(C=1.0, kernel='linear', degree=17, gamma='auto')
SVM.fit(x_train,y_train)
predictions_SVM = SVM.predict(x_test)
```

```
for i in range(0,905):
  print(predictions_SVM[i])
  print(y_test[i])
```

```
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predictions_SVM))
```

```
for i in range(0,905):
  if y_test[i]==9:
    y_test[i]=1
  else:
    y_test[i]=0
```

```
for i in range(0,905):
  if predictions_SVM[i]==9:
    predictions_SVM[i]=1
  else:
    predictions_SVM[i]=0
```

```
from sklearn.metrics import precision_score, recall_score
print(precision_score(y_test,predictions_SVM))
print(recall_score(y_test,predictions_SVM))
```
