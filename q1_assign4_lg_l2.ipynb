```
import pandas as pd

df = pd.read_csv('/content/adult.data',sep=",")
```

```
print(df)
```

```
print(df)
target=df.iloc[:,-1]
print(target)
```

```
print(target)
```

```
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
#from sklearn.cross_validation import train_test_split

X = df.values.tolist()
Y=target
x_train = []
y_train = []
x_test = []
y_test = []

x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.8,shuffle=True)
```

```

print(target)
```

```
import numpy as np
x_train = np.array(x_train)
y_train = np.array(y_train)
x_test = np.array(x_test)
y_test = np.array(y_test)
```

```
y_train = y_train.reshape(26048,1)
y_test = y_test.reshape(6512,1)
```

```
from sklearn.preprocessing import LabelEncoder
Le = LabelEncoder()
for i in range(0, 26048):
  x_train[i]= Le.fit_transform(x_train[i])
for i in range(0, 6512):
  x_test[i] = Le.fit_transform(x_test[i])
```

```
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
ohe.fit(x_train)
x_train = ohe.transform(x_train)
x_test = ohe.transform(x_test)
```

```
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test)
```

```
y_train=y_train.astype('str')
y_test=y_test.astype('str')
```

```
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression(penalty='l2',C=1.0)
lg.fit(x_train,y_train)
predictions_lg = lg.predict(x_test)
```

```
for i in range(0,6512):
  print(predictions_lg[i])
  print(y_test[i])
```

```
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predictions_lg))
```

```
for i in range(0,6512):
  if y_test[i]=='>50K':
    y_test[i]=1
  else:
    y_test[i]=0
```

```
for i in range(0,6512):
  if predictions_lg[i]=='>50K':
    predictions_lg[i]=1
  else:
    predictions_lg[i]=0
```

```
from sklearn.metrics import precision_score, recall_score
print(precision_score(y_test,predictions_lg))
print(recall_score(y_test,predictions_lg))
```
